```python
# df_setosa = df_setosa.reset_index(drop=True)
q25 = df_setosa['SepalLengthCm'].quantile(0.25)
q75=df_setosa['SepalLengthCm'].quantile(0.75)

iqr = q75 - q25

lower_bound = q25 - 1.5 * iqr
upper_bound = q75 + 1.5 * iqr

filtered_df = df_setosa[(df_setosa['SepalLengthCm'] < lower_bound) | (df_setosa['SepalLengthCm'] > upper_bound) ]
count=filtered_df['SepalLengthCm'].count()
print(count)

#it shows 0 columns
#it means it species setosa has 0 outliers in sepalLength
```
